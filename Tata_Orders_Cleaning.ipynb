```python
import pandas as pd
df = pd.read_csv("Online Retail.csv")
```

```python
df.head(20)
```

```python
#checking for negative values in quality column
df['Quantity'].unique()
```

```python
# Filter the rows based on the conditions
df_cleaned = df[(df['Quantity'] >= 1) & (df['UnitPrice'] >= 0)]
```

```python
# Save the cleaned dataframe to a new CSV file
cleaned_file_name = 'cleaned_data.csv'
df_cleaned.to_csv(cleaned_file_name, index=False)

# Allow user to download the cleaned file
files.download(cleaned_file_name)
```
